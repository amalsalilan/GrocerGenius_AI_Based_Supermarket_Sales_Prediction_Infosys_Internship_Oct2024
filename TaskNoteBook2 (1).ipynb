```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example DataFrame (replace this with  actual DataFrame)
data = pd.read_csv('Train.csv')  

# Continuous numerical features
continuous_columns = ['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Item_Outlet_Sales']

# Z-score threshold
z_threshold = 3 # usually kept around + - 3

# Loop through each continuous column
for col in continuous_columns:
    # Calculate Z-scores
    z_scores = np.abs((data[col] - data[col].mean()) / data[col].std())
    
    # Identify outliers
    outliers = z_scores > z_threshold
    num_outliers = np.sum(outliers)  # Count outliers
    print(f"{col}: {num_outliers} outliers detected.")
    
    # Replace outliers with the median
    data[col][outliers] = data[col].median()

# Plot boxplots for each column after treating outliers
plt.figure(figsize=(15, 10))
for i, col in enumerate(continuous_columns, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col} after Outlier Treatment')

plt.tight_layout()
plt.show()
```

```python
#Target Variable transforamtion
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
# Example of loading your dataset
df = pd.read_csv('Train.csv')  

# Ensure you have the target variable in your dataset
target_variable = 'Item_Outlet_Sales'
plt.figure(figsize=(12, 6))
sns.histplot(df[target_variable], bins=30, kde=True)
plt.title('Original Target Variable Distribution')
plt.xlabel('Item Outlet Sales')
plt.ylabel('Frequency')
plt.show()



# Log Transformation
df['Item_Outlet_Sales_Log'] = np.log(df[target_variable] + 1)  # Adding 1 to avoid log(0)

plt.figure(figsize=(12, 6))
sns.histplot(df['Item_Outlet_Sales_Log'], bins=30, kde=True)
plt.title('Log Transformed Target Variable Distribution')
plt.xlabel('Log(Item Outlet Sales + 1)')
plt.ylabel('Frequency')
plt.show()


#

# Square Root Transformation
df['Item_Outlet_Sales_Sqrt'] = np.sqrt(df[target_variable])

plt.figure(figsize=(12, 6))
sns.histplot(df['Item_Outlet_Sales_Sqrt'], bins=30, kde=True)
plt.title('Square Root Transformed Target Variable Distribution')
plt.xlabel('Square Root(Item Outlet Sales)')
plt.ylabel('Frequency')
plt.show()


# 

# Box-Cox Transformation
df['Item_Outlet_Sales_BoxCox'], _ = stats.boxcox(df[target_variable] + 1)  # Shift to make positive

plt.figure(figsize=(12, 6))
sns.histplot(df['Item_Outlet_Sales_BoxCox'], bins=30, kde=True)
plt.title('Box-Cox Transformed Target Variable Distribution')
plt.xlabel('Box-Cox(Item Outlet Sales + 1)')
plt.ylabel('Frequency')
plt.show()



```
