### Conditional Based Selection

```python
#return all rows where the "country" column is "Italy in true or false
data.country=="Italy"
```

```python
#  returns a new DataFrame containing only the rows where "country" is "Italy".
data.loc[data.country=="Italy"]
```

```python
#results in DataFrame with rows where "country" is "Italy" and "points" are 90 or higher.
data.loc[(data.country =="Italy") & (data.points>=90)]
```

```python
#results in DataFrame with rows where "country" is "Italy" or "points" are 90 or higher.
data.loc[(data.country=="Italy") | (data.points>=100)]
```

```python
# shows all rows where the "country" is either "Italy" or "France".
data.loc[data.country.isin(["Italy","France"])]
```
