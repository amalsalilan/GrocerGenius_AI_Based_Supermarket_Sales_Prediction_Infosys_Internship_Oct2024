Now we will check for the missing values in the dataset using the commands isnull()

```python
df.isnull().sum()
```

```python
df['Item_Weight'].fillna(df['Item_Weight'].mean(), inplace=True)
#this helps us to fill the na values with the mean of the data
#ORRR we can do the same with the median data also
#df['column_name'].fillna(df['column_name'].median(), inplace=True)
df
```

```python
df.isnull().sum()
#Now we have succ handeled the missing datas in the Item_Weight Col 
```

```python
# Now to handel the data in the Outlet_Size column we need to use the mode data as the values in it are non numeric
```

```python
df['Outlet_Size'].fillna(df['Outlet_Size'].mode()[0], inplace=True)
df
```

```python
df.isnull().sum()
```

```python
import pandas as pd

df = pd.read_csv('Train.csv')
print(df.shape)      # Get rows and columns count
print(df.info())     # Data types and non-null counts
print(df.head())     # First few rows to get an overview
```

```python
print(df.isnull().sum())  # Summing up all missing values per column
```

```python
#for visuals i m using heatma[
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.show()
```

```python
#it shows there are a lot of missing datas in the Outlet_Size and Item weight :::
```

```python
print(df.describe())  # Numerical stats
print(df['Item_Weight'].value_counts())  # Categorical counts
#we use describe to get count , mean , median and every other thing down below
#we can count category wise too by using value_count 
```

```python
df['Item_Weight'].hist(bins=20)
plt.show()
#histograms to check the distribution of datas
```

```python
#box plot for outliers
sns.boxplot(x='Outlet_Size', y='Item_Weight', data=df)
plt.show()
```

```python
sns.boxplot(x='Outlet_Type', y='Item_Outlet_Sales', data=df)
plt.show()
```

```python
#We got the ouytliers here , the valuses which are distinctive than the other values present in the data set columns
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Item_MRP', y='Item_Outlet_Sales', data=df)
plt.title('Relationship Between Item MRP and Sales')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df.boxplot(column='Item_Visibility', by='Outlet_Type')
plt.title('Item Visibility Distribution by Outlet Type')
plt.suptitle('')  # Suppress default title
plt.xlabel('Outlet Type')
plt.ylabel('Item Visibility')
plt.xticks(rotation=90)
plt.show()
```

```python
df['Outlet_Age'] = 2024 - df['Outlet_Establishment_Year']
plt.figure(figsize=(10, 6))
sns.lineplot(x='Outlet_Age', y='Item_Outlet_Sales', data=df, marker='o')
plt.title('Sales Performance by Outlet Age')
plt.show()



#New column fro outlet_age
```

```python
plt.bar(df['Outlet_Age'], df['Item_Outlet_Sales'])
plt.title('Impact of Outlet Age on Sales')
plt.xlabel('Outlet Age')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
correlation_matrix = df[['Item_Weight', 'Item_MRP', 'Outlet_Age', 'Item_Outlet_Sales']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Item_Visibility', y='Item_Outlet_Sales', data=df)
plt.title('Sales vs Item Visibility')
plt.xlabel('Item Visibility')
plt.ylabel('Sales')
plt.show()
```

```python
top_products = df.groupby('Item_Identifier')['Item_Outlet_Sales'].sum().nlargest(10).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Item_Identifier', y='Item_Outlet_Sales', data=top_products)
plt.title('Top 10 Products by Sales')
plt.xlabel('Item Identifier')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
top_outlets = df.groupby('Outlet_Identifier')['Item_Outlet_Sales'].sum().nlargest(10).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Outlet_Identifier', y='Item_Outlet_Sales', data=top_outlets)
plt.title('Top 10 Performing Outlets')
plt.xlabel('Outlet Identifier')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Item_Type', y='Item_Weight', data=df)
plt.xticks(rotation=90)
plt.title('Item Weight Distribution by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Item Weight')
plt.show()
```
