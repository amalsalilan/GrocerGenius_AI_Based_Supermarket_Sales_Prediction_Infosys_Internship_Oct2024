## **Handling Missing Values**

```python
values_missed = df.isnull().sum()
```

```python
values_missed
```

```python
## Filling up the missing values using Mean
df['Item_Weight'] = df.groupby('Item_Type')['Item_Weight'].transform(lambda x: x.fillna(x.mean()))
```

```python
## FIlling up missing values using Mode
df['Outlet_Size'] = df.groupby('Outlet_Type')['Outlet_Size'].transform(lambda x: x.fillna(x.mode()[0]))
```

```python
# checking again for null values
df.isnull().sum()
```

## *Observation - No Missing values found*

```python
import matplotlib.pyplot as plt
import seaborn as sns
 # Seaborn used for Statistical Data Visualisation
```

## 1. Sales distribution by Item Type

```python
item_sales = df.groupby('Item_Type')['Item_Outlet_Sales'].sum().sort_values(ascending=False)
print("\nSales by Item Type:\n", item_sales)
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x=item_sales.index, y=item_sales.values)
plt.xticks(rotation=90)
plt.title('Sales Distribution by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.show()
```

Observation:

```python
least_sold = item_sales.idxmin()
highest_sold = item_sales.idxmax()

print(f"'{highest_sold}' is the top sold item and '{least_sold}' is the least sold item.")
```

## 2. Sales Performance by Outlet Establishment Year

```python
annual_sales = df.groupby('Outlet_Establishment_Year')['Item_Outlet_Sales'].mean().sort_values()
print("\nThe Average Sales by Outlet Establishment Year is given below:\n", annual_sales)
```

```python
# Visualisation
```

```python
plt.figure(figsize=(10,7))
sns.lineplot(x=annual_sales.index, y=annual_sales.values,color = 'black',marker = 'o' )
plt.title('Sales Performance by Outlet Establishment Year')
plt.xlabel('Year of Establishment')
plt.ylabel('Avg Sales')
plt.show()
```

Observation

```python
highest = annual_sales.idxmax()
lowest = annual_sales.idxmin()
print(f" Outlets established in '{highest}' holds the highest average sales and outlest established in '{lowest}' has the lowest average sales")
```

# 3. Sales Distribution Across Outlet Types

```python
sales_of_outlets = df.groupby('Outlet_Type')['Item_Outlet_Sales'].describe()
sales_of_outlets
```

Visualisation

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='Outlet_Type', y='Item_Outlet_Sales', data = df)
plt.title('Sales Distribution Across various Outlet Types')
plt.xlabel('Outlet Type')
plt.ylabel('Annual Sales')
plt.show()
```

Observation

```python
high_sale_outlet = df.groupby('Outlet_Type')['Item_Outlet_Sales'].mean().idxmax()
low_sale_outlet = df.groupby('Outlet_Type')['Item_Outlet_Sales'].mean().idxmin()
print(f"The '{high_sale_outlet}' outlets have the highest average sales and '{low_sale_outlet}' have the lowest average sales.")
```

```python
## 4. Effect of Outlet Size on Sales
```

```python
eff_sales = df.groupby('Outlet_Size')['Item_Outlet_Sales'].mean().sort_values(ascending=False)
print("\nAverage Sales by Outlet Size:\n", eff_sales)
```

Visulaisation

```python
plt.figure(figsize=(10,6))
sns.barplot(x=eff_sales.index, y=eff_sales.values)
plt.title('Sales effected by outlet size')
plt.xlabel('Size of Outlet')
plt.ylabel('Avg Sales of outlets')
plt.show()
```

Observation: 
Medium sized outlets have the higest effect on avg sales
